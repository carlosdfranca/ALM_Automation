## Pegando os dados da carteira

```python
nav.get('https://siru.com.br/plataformarpsis/Usuarios/uCarteira.aspx')

# https://www.softwaretestingmaterial.com/how-to-handle-web-tables-in-selenium-python/
# to identify the table rows
r = nav.find_elements(By.XPATH, '//*[@id="DataTables_Table_0"]/tbody/tr')
# to identify table columns
c = nav.find_elements(By.XPATH, '//*[@id="DataTables_Table_0"]/tbody/tr[6]/td')
# to get row count with len method
rc = len (r)
# to get column count with len method
cc = len (c)

dict_fundos_carteira = {}

# to traverse through the table rows excluding headers
for i in range (1, rc + 1) :
    # to traverse through the table column
    linha = []
    for j in range (1, cc + 1) :
        # to get all the cell data with text method
        d = nav.find_element(By.XPATH, "//tr["+str(i)+"]/td["+str(j)+"]").text
        linha.append(d)
        if j == 8:
            if linha[0] != 'TÍTULOS PÚBLICOS':
                dict_fundos_carteira[linha[0]] = linha[1], linha[2], linha[3], linha[4], linha[5], linha[6], linha[7]     

print(dict_fundos_carteira)
```

```python
df_fundos_carteira = pd.DataFrame.from_dict(dict_fundos_carteira, orient='index').reset_index().rename(columns={'index':'Fundo', 0:'Disponibilidade Financeira', 1:'Carência', 2:'Saldo', 3:'Particíp. S/ Total', 4:'Qtd Cotistas', 5:'% S/ PL do Fundo', 6:'Enquadramento'})

display(df_fundos_carteira)
```
